```python
import numpy as np
import pandas as pd
df = pd.read_csv('customer.csv')
df.sample(5)
```

```python
df = df.iloc[:,2:]
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,0:2], df.iloc[:,-1], test_size=0.2)
X_train
```
