### 1. Scatterplot (Numerical - Numerical)

```python
sns.scatterplot(x='total_bill', y='tip', hue='sex', style='smoker', size='size', data=tips)
```

### 2. Bar Plot (Numerical - Categorical)

```python
titanic.head()
```

```python
sns.barplot(x='Pclass', y='Fare', hue='Sex', data=titanic)
```

### 3. Box Plot (Numerical - Categorical)

```python
sns.boxplot(x='Age', y='Sex', hue='Survived', data=titanic)
```

### 4. Distplot (Numerical - Categorical)

```python
sns.displot(data=titanic, x='Age', hue='Survived', multiple="stack", kde=True)
```

```python
sns.kdeplot(titanic[titanic['Survived'] == 0]['Age'], label='Not Survived', fill=True)
sns.kdeplot(titanic[titanic['Survived'] == 1]['Age'], label='Not Survived', fill=True)
```

### 5. HeatMap (Categorical - Categorical)

```python
titanic.head(3)
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'], titanic['Survived']))
```

```python
titanic.groupby('Pclass')['Survived'].mean() * 100
```

### 6. ClusterMap (Categorical - Categorical)

### 6. ClusterMap (Categorical - Categorical)

```python
pd.crosstab(titanic['SibSp'], titanic['Survived'])
```

```python
sns.clustermap(pd.crosstab(titanic['Parch'],titanic['Survived']))
```
